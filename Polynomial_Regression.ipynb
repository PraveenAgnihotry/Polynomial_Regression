```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<h3> Reading the Data

```python
# Reading the excel data
df_raw = pd.read_excel('Data Take Home Assignment 1 Exercise A.xlsx', sheet_name='Sheet5')
```

```python
# Selecting the data according to our group number 6
group_no = 6

df = df_raw[((group_no-1)*20+2) : (group_no*20+1)+1]
df
```

```python
# Info about the dataset
df.describe()
```

```python
df = df.sort_values(by=['X'])
```

```python
# Defining X and Y array values
X = df['X'].values # Temperature Column
Y = df['Y'].values # Net hourly electrical energy output Column
print("X: ",X)
print("Y: ",Y)
```

```python
# Plot of all the data points on a graph
plt.scatter(X, Y)
plt.show()
```

<h3>Data Pre-processing

```python
# Normalizing the input data
X = (X - np.mean(X)) / np.std(X)  # Z score normalization (mean = 0, std = 1)
```

```python
# Length of data
n = len(X)
```

```python
# Column of ones for theta_0 value, 
# X for theta_1 values and 
# X^2 values for theta_2, of the regression model equation [(theta_2 . x^2) + (theta_1 . x) + (theta_0 . 1)]
X = np.transpose(np.array([np.ones(n), X, X**2]))
X
```

```python
# Initializing the theta parameter values (theta_0, theta_1, theta_2)
theta = np.zeros(3)
```

<h3>Polynomial Regression

```python
# Polynomial regression function
def poly_reg_function(X, theta):
    '''This function returns the regression model output with the current theta parameter values'''
    X_dot_theta = np.dot(X, theta)   # represents the Regression model: [(theta_0 . 1) + (theta_1 . x) + (theta_2 . x^2)]
    return X_dot_theta
```

```python
# Defining the cost function with the power of 4
def cost_function(X, Y, theta):
    '''This func returns the cost value for the current parameters'''
    cost_value = (1 / (4 * n)) * np.sum((Y - poly_reg_function(X, theta)) ** 4)   # Cost function as defined in the assignment
    return cost_value
```

```python
# Defining the gradient descent function
def gradient_descent(X, Y, theta, alpha, iterations):
    '''This func finds the gradient/slope of the equation and updates the parameter values to a more ideal value 
    such that it reduces the cost function value'''
    cost_history = []
    theta_history = []
    
    for i in range(iterations):
        predictions = poly_reg_function(X, theta)   # predictions with the current values of theta
        errors = predictions - Y   # Amount of error compared to the actual value
        
        # Gradient computation
        grad = (1 / n) * np.dot(np.transpose(X), (errors ** 3))  # Derivative of cost function with power of 4
        
        grad = np.clip(grad, -20, 20) # Clip the gradients between -20 and 20.
        # Used because the errors are being cubed thus leading to massive values of gradient
        
        # Updates theta parameters
        theta = theta - (alpha * grad) # Gives the updated values of theta_0, theta_1 and theta_2
        # here alpha is the learning rate which controls the step sizes
        theta_history.append(theta)
        
        # Computes and stores the latest cost value with the new parameters
        cost = cost_function(X, Y, theta)
        cost_history.append(cost)
    
    return theta, cost_history, theta_history
```

```python
# Parameters for gradient descent (Values found by trial and error)
alpha = 0.01  # Defines the Learning rate
iterations = 5000  # Number of iterations to undergo
```

```python
# Performing gradient descent to optimize theta
theta_optimal, cost_history, theta_history = gradient_descent(X, Y, theta, alpha, iterations)
```

```python
# Output the optimized parameters
print("Optimized theta parameters:", theta_optimal)
print("Historical costs:", cost_history[-10:])
```

<h3>Plots

```python
# Making predictions with the final theta values
y_pred = poly_reg_function(X, theta_optimal)
y_pred
```

```python
# Plotting the original values vs the final prediction function
plt.scatter(df['X'], Y, label='Actual Energy Output')
plt.plot(df['X'], y_pred, label='Predicted Energy Output', color='red') # Final Prediction line
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Net hourly electrical energy output (MW)')
plt.legend()
plt.show()
```

```python
# Plotting the original values vs every 50 theta value showing the improvement in the parameter values
import matplotlib.cm as cm

plt.scatter(df['X'], Y, label='Actual Energy Outputs', linewidth=0.2)

# To generate a color map
cmap = plt.cm.plasma  # using colormaps
colors = cmap(np.linspace(0, 1, 200))  # 100 evenly spaced colors

for i in range(200):
    plt.plot(df['X'], poly_reg_function(X, theta_history[i*25]), linewidth=0.5, color=colors[i])
    
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Net hourly electrical energy output (MW)')
plt.legend()
plt.show()
```
